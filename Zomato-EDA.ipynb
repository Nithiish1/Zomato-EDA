# Zomato Data Set Analysis and Visualisation


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("dark_background")
```

 # Reading CSV

```python
data = pd.read_csv("zomato.csv")
```

```python
data.head(3)
```

```python
data.shape
```

```python
data.columns
```

```python
 data = data.drop(["url","phone","address","menu_item","dish_liked"],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python
data.rate.unique()
```

```python
data.rate.isnull().sum()
```

```python
def handle(value):
    if(value=='NEW' or value=="-"):
        return np.nan
    else:
        value = str(value).split("/")
        value=value[0]
        return float(value)
```

```python
data.rate = data.rate.apply(handle)
data.rate.unique()
```

```python
data.rate.isnull().sum()
```

```python
data.rate.fillna(data.rate.mean(),inplace=True)
data.rate.isnull().sum()
```

```python
data.info()
data.shape
```

# Droping NULL Values

```python
data.dropna(inplace=True)
data.isnull().sum()
data.shape
```

```python
data.head(3)
```

```python
data.rename(columns={"approx_cost(for two people)":"cost2people","listed_in(type)":"type","listed_in(city)":"city"},inplace=True)
```

```python
data.shape
```

```python
data.location.unique()
```

```python
data.city.unique()
```

```python
data = data.drop("city",axis=1)
```

```python
data
```

```python
data.cost2people.unique()
```

# Remove , from cost2people

```python
def hand(value):
    value=str(value)
    if "," in value:
        value = value.replace(",","")
        return int(value)
    else:
        return float(value)
```

```python
data.cost2people = data.cost2people.apply(hand)
```

```python
data.cost2people.unique()
```

```python
data.head(3)
```

# Cleaning rest_type Column

```python
data.rest_type.unique()
```

```python
data.rest_type.value_counts()
```

```python
rest_types  = data.rest_type.value_counts()
```

```python
rest_types_counts = rest_types[rest_types<1000]
rest_types_counts
```

```python
def handles(value):
    if(value in rest_types_counts):
        return "others"
    else:
        return value
data.rest_type = data.rest_type.apply(handles)
```

```python
data
```

```python
data.rest_type.unique()
```

# Cleaning Location Column

```python
pd.set_option("display.max_rows",100)
location=data.location.value_counts()
```

```python
location.values
```

```python
location_lessthan250 = location[location.values<250]
def handle_location(value):
    if(value in location_lessthan250):
        return "others"
    else:
        return value
data.location = data.location.apply(handle_location)
```

```python
data.location.unique()
```

```python
data.location.value_counts()
```

# Cleaning Cusines Column

```python
cuisines=data.cuisines.value_counts()
cuisines
```

```python
pd.set_option("display.max_rows",100)
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
data['cuisines'] = data['cuisines'].apply(handle_cuisines)
data['cuisines'].value_counts()
```

# Visualizing Online Order Facility,Location Wise

```python
df2 = data.groupby(['book_table',"location"])['name'].count()
df2.to_csv('location_booktable.csv')
df2 = pd.read_csv('location_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['location'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind="bar",figsize=(15,8))
```
